```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data1 = pd.read_csv('RESULTS_sobolsampling-2048.csv')
data2 = pd.read_csv('RESULTS_sobolsampling-2048-2.csv')

data1 = data1.drop(data1.columns[0], axis=1)
data2 = data2.drop(data2.columns[0], axis=1)

colnames = data1.iloc[0:6,0]
#print(f'{colnames}')
cols = []
for i in range(0,5):
    #print(f'{colnames[i]}')
    cols.append(colnames[i])
#print(f'{cols}')

data = []
for i in range(1, data1.shape[1]):
    d = data1.iloc[0:5,i]
    #print(f'd:\n{d}')
    tmp = []
    for j in range(0,5):
        tmp.append(float(d[j]))
    data.append(tmp)    
    #if i == 2:
    #    break

#print(f'{data}')
data1 = pd.DataFrame(np.array(data), columns=[cols])

colnames = data2.iloc[0:6,0]
#print(f'{colnames}')
cols = []
for i in range(0,5):
    #print(f'{colnames[i]}')
    cols.append(colnames[i])
#print(f'{cols}')

data = []
for i in range(1, data2.shape[1]):
    d = data2.iloc[0:5,i]
    #print(f'd:\n{d}')
    tmp = []
    for j in range(0,5):
        tmp.append(float(d[j]))
    data.append(tmp)    
    #if i == 2:
    #    break

#print(f'{data}')
data2 = pd.DataFrame(np.array(data), columns=[cols])

print(f'{data1}')
print(f'{data2}')
```

## plots

```python
X1 = data1.SigC
Y1 = data1.SigH
X2 = data2.SigC
Y2 = data2.SigH

plt.figure()
sc1 = plt.scatter(X1, Y1, color='blue', label="Sobol Sampling Dataset 1")
sc2 = plt.scatter(X2, Y2, color='orange', label="Sobol Sampling Dataset 2")
plt.xlabel("Sigma C [unit]")
plt.ylabel("Sigma H [unit]")
plt.title("Sobol Sampling Datasets")
plt.legend()
plt.savefig('Compare_Sobolsampling_datasets_SigC-SigH.png')


X1 = data1.SigC
Y1 = data1.EpsC
X2 = data2.SigC
Y2 = data2.EpsC

plt.figure()
sc1 = plt.scatter(X1, Y1, color='blue', label="Sobol Sampling Dataset 1")
sc2 = plt.scatter(X2, Y2, color='orange', label="Sobol Sampling Dataset 2")
plt.xlabel("Sigma C [unit]")
plt.ylabel("Epsilon C [unit]")
plt.title("Sobol Sampling Datasets")
plt.legend()
plt.savefig('Compare_Sobolsampling_datasets_SigC-EpsC.png')


X1 = data1.SigC
Y1 = data1.EpsH
X2 = data2.SigC
Y2 = data2.EpsH

plt.figure()
sc1 = plt.scatter(X1, Y1, color='blue', label="Sobol Sampling Dataset 1")
sc2 = plt.scatter(X2, Y2, color='orange', label="Sobol Sampling Dataset 2")
plt.xlabel("Sigma C [unit]")
plt.ylabel("Epsilon H [unit]")
plt.title("Sobol Sampling Datasets")
plt.legend()
plt.savefig('Compare_Sobolsampling_datasets_SigC-EpsH.png')


X1 = data1.SigH
Y1 = data1.EpsC
X2 = data2.SigH
Y2 = data2.EpsC

plt.figure()
sc1 = plt.scatter(X1, Y1, color='blue', label="Sobol Sampling Dataset 1")
sc2 = plt.scatter(X2, Y2, color='orange', label="Sobol Sampling Dataset 2")
plt.xlabel("Sigma H [unit]")
plt.ylabel("Epsilon C [unit]")
plt.title("Sobol Sampling Datasets")
plt.legend()
plt.savefig('Compare_Sobolsampling_datasets_SigH-EpsC.png')


X1 = data1.SigH
Y1 = data1.EpsH
X2 = data2.SigH
Y2 = data2.EpsH

plt.figure()
sc1 = plt.scatter(X1, Y1, color='blue', label="Sobol Sampling Dataset 1")
sc2 = plt.scatter(X2, Y2, color='orange', label="Sobol Sampling Dataset 2")
plt.xlabel("Sigma H [unit]")
plt.ylabel("Epsilon H [unit]")
plt.title("Sobol Sampling Datasets")
plt.legend()
plt.savefig('Compare_Sobolsampling_datasets_SigH-EpsH.png')


X1 = data1.EpsC
Y1 = data1.EpsH
X2 = data2.EpsC
Y2 = data2.EpsH

plt.figure()
sc1 = plt.scatter(X1, Y1, color='blue', label="Sobol Sampling Dataset 1")
sc2 = plt.scatter(X2, Y2, color='orange', label="Sobol Sampling Dataset 2")
plt.xlabel("Epsilon C [unit]")
plt.ylabel("Epsilon H [unit]")
plt.title("Sobol Sampling Datasets")
plt.legend()
plt.savefig('Compare_Sobolsampling_datasets_EpsC-EpsH.png')
```
