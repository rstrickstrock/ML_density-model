# n-Octane Relative Conformational Energies - Neural Network Predictor

## import libraries

```python
# Torch
import torch
from torch.utils.data import TensorDataset, DataLoader
import torch.nn.functional as F

# for data preparation
import optuna
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# other stuff
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

## load training data

```python
df_data_training = pd.read_csv("RESULTS_Training-2401.csv")
#print(f'{df_data_training}')
df_data_training = df_data_training.drop(columns='Unnamed: 0')
#print(f'{df_data_training}')

colnames_input = df_data_training.iloc[0:4,0].to_numpy()
#print(f'{colnames_input}')
cols = []
for i in range(0,len(colnames_input)):
    #print(f'{colnames_input[i]}')
    cols.append(colnames_input[i])
#print(f'{cols}')

colnames_output = df_data_training.iloc[6:,0].to_numpy()
#print(f'{colnames_output}')
for i in range(0,len(colnames_output)):
    #print(f'{colnames_output[i]}')
    cols.append(colnames_output[i])
#print(f'{cols}')

data = []
for i in range(1, df_data_training.shape[1]):
    d = df_data_training.iloc[0:len(colnames_input),i].to_numpy()
    tmp = []
    for j in range(0,len(colnames_input)):
        tmp.append(d[j])
    d = df_data_training.iloc[6:,i].to_numpy()
    for j in range(0,len(colnames_output)):
        tmp.append(float(d[j]))
    data.append(tmp)    
    #if i == 2:
    #    break

#print(f'{data}')
df_data_training = pd.DataFrame(np.array(data), columns=[cols])
#print(f'{df_data_training}')
```

## load testing data

```python
df_data_testing = pd.read_csv("RESULTS_Testing-1296.csv")
#print(f'{df_data_testing}')
#print(f'{df_data_testing.columns}')
df_data_testing = df_data_testing.drop(columns='Unnamed: 0')
#print(f'{df_data_testing}')
#print(f'{df_data_testing.columns}')
#print(f'{df_data_testing.shape[1]}')

colnames_input = df_data_testing.iloc[0:4,0].to_numpy()
#print(f'{colnames_input}')
cols = []
for i in range(0,len(colnames_input)):
    #print(f'{colnames_input[i]}')
    cols.append(colnames_input[i])
#print(f'{cols}')

colnames_output = df_data_testing.iloc[6:,0].to_numpy()
#print(f'{colnames_output}')
for i in range(0,len(colnames_output)):
    #print(f'{colnames_output[i]}')
    cols.append(colnames_output[i])
#print(f'{cols}')

data = []
for i in range(1, df_data_testing.shape[1]):
    d = df_data_testing.iloc[0:len(colnames_input),i].to_numpy()
    tmp = []
    for j in range(0,len(colnames_input)):
        tmp.append(d[j])
    d = df_data_testing.iloc[6:,i].to_numpy()
    for j in range(0,len(colnames_output)):
        tmp.append(float(d[j]))
    data.append(tmp)    
    #if i == 2:
    #    break

#print(f'{data}')
df_data_testing = pd.DataFrame(np.array(data), columns=[cols])
#print(f'{df_data_testing}') 

#print(f'{data}')
df_data_training = pd.DataFrame(np.array(data), columns=[cols])
print(f'{df_data_training}')
```

## prepare training data

```python
df_x_train = df_data_training[['SigC', 'SigH', 'EpsC', 'EpsH']]
x_train = torch.Tensor(df_x_train.to_numpy().astype(float))
#print(x_train)
df_y_train = df_data_training.drop(columns=['SigC', 'SigH', 'EpsC', 'EpsH'])
#print(df_y_train)
y_train = torch.Tensor(df_y_train.to_numpy().astype(float))
#print(y_train)

train_data = TensorDataset(x_train, y_train)

# setting batch size and dataloader
batch_size = len(x_train)
train_dl = DataLoader(train_data, batch_size, shuffle=True)
```

## prepare testing data

```python
df_x_test = df_data_testing[['SigC', 'SigH', 'EpsC', 'EpsH']]
x_test = torch.Tensor(df_x_test.to_numpy().astype(float))

df_y_test = df_data_testing.drop(columns=['SigC', 'SigH', 'EpsC', 'EpsH'])
y_test = torch.Tensor(df_y_test.to_numpy().astype(float))

test_data = TensorDataset(x_test, y_test)
```

## a) initialize model and create structure

```python
model = net = torch.nn.Sequential(
        torch.nn.Linear(4, 512),
        torch.nn.LeakyReLU(),
        torch.nn.Linear(512, 2048),
        torch.nn.LeakyReLU(),
        torch.nn.Linear(2048, 8192),
        torch.nn.LeakyReLU(),
        torch.nn.Linear(8192, 2048),
        torch.nn.LeakyReLU(),
        torch.nn.Linear(2048, 512),
        torch.nn.LeakyReLU(),
        torch.nn.Linear(512, 96),
    )
model
```

### model learning

```python
epoch_count = 0
num_epochs = 500

loss_fn = torch.nn.MSELoss()
# optimizer = torch.optim.Adam(model.parameters(), lr=lr_optuna)
optimizer = torch.optim.Adam(model.parameters(), lr=0.002)

for epoch in range(num_epochs):
    for step, (xb, yb) in enumerate(train_dl):
        pred = model(xb)

        loss = loss_fn(pred, yb)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    # if epoch%10 == 9:
    print('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, num_epochs, loss.item()))
    epoch_count = epoch_count+1
```

```python
#print(f"Total epoch count: {epoch_count}")
#print()

loss_fn = torch.nn.MSELoss()

predictions = model(x_test)
#print(f'{predictions}')
rmse = torch.sqrt(loss_fn(predictions, y_test)).item()
predictions_np = predictions.detach().numpy()
r2 = r2_score(y_test, predictions_np)

# Print the final predictions
print(f'RMSE = {rmse}, R^2 = {r2}')
```

### calculate rmse

```python
def rmse_loss(target, calc):
    #print(f'{target}\n{calc}')
    try:
        target = np.array(target)
    except:
        print("1")
        
    try:
        calc = np.array(calc)
    except:
        print("2")
    
    zero_idx = []
    for i in range(0, len(target)):
        if target[i] == 0:
            target[i] = 10e-1
            if calc[i] == 0:
                calc[i] = calc[i] + 10e-1
                zero_idx.append(i)
            
        
    loss = ((target - calc)/target)
    #print(f'{loss}')
    loss = np.square(loss)
    #print(f'{loss}')
    loss = np.sqrt(loss)
    #print(f'{loss}')
    n = len(loss)
    #print(f'{n}')
    for i in zero_idx:
        loss[i] = 0.0
    rmse_loss = np.sum(loss)
    #print(f'{rmse_loss}')
    rmse_loss = rmse_loss/n
    #print(f'{rmse_loss}')
    return rmse_loss

test_rmse = rmse_loss(target_energies_sorted, target_energies_sorted)
print(f'Test: {test_rmse} (should be 0.0!)')
test_rmse = rmse_loss(prediction_sorted, prediction_sorted)
print(f'Test: {test_rmse} (should be 0.0!)')
test_rmse = rmse_loss(prediction_sorted, prediction)
print(f'Test: {test_rmse} (should NOT be 0.0!)')
plt.plot(x, prediction, '-x', label='Unsorted', color='#069AF3')
plt.plot(x, prediction_sorted, '-x', label='Sorted', color='#F97306')
plt.legend()
plt.show()

test_rmse = rmse_loss(target_energies_sorted, prediction)
print(f'Target_sordet vs Pred: {test_rmse}')


rmse_pred = rmse_loss(target_energies_sorted, prediction_sorted)
print(f'RMSE pred: {rmse_pred}')
rmse_calc = rmse_loss(target_energies_sorted, optimization_rce_results_sorted)
print(f'RMSE calc: {rmse_calc}')
```
