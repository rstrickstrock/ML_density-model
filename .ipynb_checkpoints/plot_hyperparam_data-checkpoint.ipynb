```python
df_target = pd.read_csv('density_input.txt', sep=' ', nrows=12,
                           header=None, names=['Temperature', 'Name', 'Density'])

df_opt1 = pd.read_csv('density_input.txt', sep=' ', skiprows=range(0, 12), nrows=12,
                           header=None, names=['Temperature', 'Name', 'Density', 'Error'])

df_opt2 = pd.read_csv('density_input.txt', sep=' ', skiprows=range(0, 24), nrows=12,
                           header=None, names=['Temperature', 'Name', 'Density', 'Error'])

display(df_target)
display(df_opt1)
display(df_opt2)
```

Replace substrings in `Name`:

```python
df_target['Name'] = df_target['Name'].str.replace('\d+\_', '', regex=True)
df_opt1['Name'] = df_opt1['Name'].str.replace('\d+\_', '', regex=True)
df_opt2['Name'] = df_opt2['Name'].str.replace('\d+\_', '', regex=True)

display(df_target)
display(df_opt1)
display(df_opt2)
```

Calculate the loss and loss error:

```python
df_opt1['loss'] = ((df_opt1['Density'] - df_target['Density'])/df_target['Density']).abs()*100
df_opt1['loss_error'] = (100/df_target['Density'])*df_opt1['Error']

df_opt2['loss'] = ((df_opt2['Density'] - df_target['Density'])/df_target['Density']).abs()*100
df_opt2['loss_error'] = (100/df_target['Density'])*df_opt2['Error']

display(df_opt1)
display(df_opt2)
```

Add run catogory and combine:

```python
df_opt1['Run'] = 'previous'
df_opt2['Run'] = 're-optimized'

df_opt = pd.concat([df_opt1, df_opt2], axis='rows', ignore_index=True)
display(df_opt)
```
